41. Write a program to Write function to find sum of two numbers.

```python
def sum(a, b):# func to find sum of 2 no.'s
    return a + b

#  taking input from the user
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

# Function calling
result = sum(num1, num2)
print(f"Sum of {num1} and {num2} ={result}")
```

42. Write a program to Write function to find maximum.

```python
def max(a, b):#func to find max no.
    if a > b:
        return a
    else:
        return b

#  taking input from the user
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

maximum = max(num1, num2)

print(f"The maximum number ammong {num1} ans {num2} is {maximum}")
```

43. Write a program to Write function to check prime.

```python
def is_prime(n):#func to check whether no. is prime or not
    if n <= 1:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True

num = int(input("Enter a number: "))#taking input from user
#printing if no. is prime or not
if is_prime(num):
    print(f"{num} is a prime number")
else:
    print(f"{num} is not a prime number")
```

44. Write a program to Write function to find factorial.

```python
def factorial(n):# func to find out factorial of the given number
    fact = 1

    for i in range(1, n + 1):
        fact *= i
    return fact

num = int(input("Enter a number: "))#taking input from the user
result = factorial(num)#func calling
print(f"The factorial of {num} is {result}")
```
