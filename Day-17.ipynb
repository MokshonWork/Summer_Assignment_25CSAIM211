65. Write a program to Merge arrays.

```python
def merge_arr(left_arr,right_arr):
    result = []
    i,j =0,0

    while i < len(left_arr) and j < len(right_arr):
        if left_arr[i] <=right_arr[j]:
            result.append(left_arr[i])
            i+=1

        else:
            result.append(right_arr[j])
            j+=1
        
    while i < len(left_arr):
            result.append(left_arr[i])
            i+=1


    while j < len(right_arr):
            result.append(right_arr[j])
            j+=1
    return result


arr_lf = [1, 2, 3, 4]
arr_right = [1, 1, 3, 4, 5, 6, 7]

print(merge_arr(arr_lf, arr_right))


```

66. Write a program to Union of arrays.

```python
arr1 = list(map(int, input("Enter first array elements: ").split()))
arr2 = list(map(int, input("Enter second array elements: ").split()))

union = []

for num in arr1:
    if num not in union:
        union.append(num)

for num in arr2:
    if num not in union:
        union.append(num)

print(f"First Array: {arr1}")
print(f"Second Array: {arr2}")
print(f"Union of Arrays: {union}")
```

67. Write a program to Intersection of arrays.

```python
arr1 = list(map(int, input("Enter first array elements: ").split()))
arr2 = list(map(int, input("Enter second array elements: ").split()))

intersection = []

for num in arr1:
    if num in arr2 and num not in intersection:
        intersection.append(num)

print(f"First Array: {arr1}")
print(f"Second Array: {arr2}")
print(f"Intersection of Arrays: {intersection}")
```

68. Write a program to Find common elements.

```python
arr1 = list(map(int, input("Enter first array elements: ").split()))
arr2 = list(map(int, input("Enter second array elements: ").split()))

common = []

for num in arr1:
    if num in arr2 and num not in common:
        common.append(num)

print(f"First Array: {arr1}")
print(f"Second Array: {arr2}")
print(f"Common Elements: {common}")
```
