21. Write a program to Convert decimal to binary.

```python
num = int(input("Enter a number: "))#--> input from user

# this Variable is used to store the binary representation
binary = ""

# Special case: if the number is 0, its binary form is also 0
if num == 0:
    binary = "0"
else:
    # Keep dividing the number by 2 until it becomes 0
    while num > 0:
        binary = str(num % 2) + binary
        num //= 2

print(f"Binary number:{binary}")
```

22. Write a program to Convert binary to decimal.

```python
binary = input("Enter a binary number: ")# Take binary input
decimal = 0# Storing decimal value
power = 0

for digit in binary[::-1]:# Reading digits from right to left
    decimal += int(digit) * (2 ** power)# Adding current digit's value
    power += 1#incrementing power by 1

print(f"Decimal number:{decimal}")
```

23. Write a program to Count set bits in a number.

```python
num = int(input("Enter a number: "))#--> input from user
count = 0# storing in count

while num > 0:# loop will run unitl num becomes 0
    count += num & 1
    num >>= 1

print(f"Number of set bits:{count}")
```

24. Write a program to Find x^n without pow().

```python
x = int(input("Enter the base (x): "))# Input base and exponent
n = int(input("Enter the exponent (n): "))

result = 1

# Multiplying x by itself n numbers of times
for i in range(n):
    result *= x

print(f"{x} ^ {n}= {result}")
```
