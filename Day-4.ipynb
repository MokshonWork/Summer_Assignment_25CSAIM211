16. Write a program to Print Armstrong numbers in a range.

```python
strt = int(input("Enter start: ")) # taking input from user for string point
end = int(input("Enter end: "))# taking input from the user for ending point 

print(f" The Armstrong no. between {strt} and {end} are: ")

for num in range(strt, end + 1):# loop will run from strt ---> end
    temp = num # storing the original value in variable name temp
    count = len(str(num))# counting no. of digits in the number
    total = 0

    while temp > 0:# normal armstrong logic
        digit = temp % 10
        total = total + digit ** count
        temp = temp // 10

    if total == num:
        print(num)
```
