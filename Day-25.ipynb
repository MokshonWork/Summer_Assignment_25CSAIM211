97. Write a program to Merge two sorted arrays.

```python
def merge_array(left, right):
    result = []
    i, j = 0, 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    while i < len(left):
        result.append(left[i])
        i += 1

    while j < len(right):
        result.append(right[j])
        j += 1

    return result

def merge_sort(arr):
    if len(arr) <=1:
        return arr
    
    mid = len(arr)//2
    left_arr = arr[:mid]
    right_arr  = arr[mid:]
    left = merge_sort(left_arr)
    right = merge_sort(right_arr)

    return merge_array(left,right)
    

nums =[3,1,2,4,1,5,2,6,4]
print(merge_sort(nums))
```

98. Write a program to Find common characters in strings.

```python
str1 = input("Enter first string: ")
str2 = input("Enter second string: ")

print(f"your input: {str1}")
print(f"your input: {str2}",end=" ")
common = ""

for ch1 in str1:
    found = False

    for ch2 in str2:
        if ch1 == ch2:
            found = True
            break

    if found:
        duplicate = False

        for ch in common:
            if ch == ch1:
                duplicate = True
                break

        if not duplicate:
            common += ch1

print(f"Common characters: {common}")
```

99. Write a program to Sort names alphabetically.

```python
n = int(input("Enter the number of names: "))

names = []

for i in range(n):
    name = input(f"Enter name {i + 1}: ")
    names.append(name)

for i in range(n):
    for j in range(0, n - i - 1):
        if names[j] > names[j + 1]:
            temp = names[j]
            names[j] = names[j + 1]
            names[j + 1] = temp

print("Names in alphabetical order:")
for name in names:
    print(name)
```

100. Write a program to Sort words by length.

```python
n = int(input("Enter the number of words: "))

words = []

for i in range(n):
    word = input(f"Enter word {i + 1}: ")
    words.append(word)

for i in range(n):
    for j in range(0, n - i - 1):

        # Find length of first word manually
        len1 = 0
        for ch in words[j]:
            len1 += 1

        # Find length of second word manually
        len2 = 0
        for ch in words[j + 1]:
            len2 += 1

        if len1 > len2:
            temp = words[j]
            words[j] = words[j + 1]
            words[j + 1] = temp

print("Words sorted by length:")
for word in words:
    print(word)
```
