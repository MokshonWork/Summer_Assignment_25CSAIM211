81. Write a program to Find string length without strlen().

```python
text = input("Enter a string: ")

count = 0

for ch in text:
    count += 1
print(f"you enter {text}")
print(f"The length of the string is: {count}")
```

82. Write a program to Reverse a string.

```python
text = input("Enter a string: ")

reverse = ""

for ch in text:
    reverse = ch + reverse

print(f"You entered: {text}")
print(f"Reversed string: {reverse}")
```

83. Write a program to Count vowels and consonants.

```python
text = input("Enter a string: ")

vowels = 0
consonants = 0

for ch in text:
    if ('A' <= ch <= 'Z') or ('a' <= ch <= 'z'):
        if ch == 'A' or ch == 'E' or ch == 'I' or ch == 'O' or ch == 'U' or \
           ch == 'a' or ch == 'e' or ch == 'i' or ch == 'o' or ch == 'u':
            vowels += 1
        else:
            consonants += 1

print(f"You entered: {text}")
print(f"Number of vowels: {vowels}")
print(f"Number of consonants: {consonants}")
```

84. Write a program to Convert lowercase to uppercase.

```python
text = input("Enter a string: ")

result = ""

for ch in text:
    if ch >= 'a' and ch <= 'z':
        result = result + chr(ord(ch) - 32)
    else:
        result = result + ch

print(f"You entered: {text}")
print(f"Uppercase: {result}")
```
