1. Write a program to Calculate sum of first N natural numbers.

```python
num = int(input("Enter the number: ")) # taking input from user for N
sum:int  = 0 # initializing variable name as zero to store sum till N

for i in range(1,num+1): # loop from 1 --> N
    sum+=i # adding all numbers till N
print(f"The Sum is {sum}") # printing sum
```

3. Write a program to Find factorial of a number.

```python
num = int(input("Enter the number: ")) # taking input from user for number
factorial:int = 1  # initializing factorial variable as 1 as we know factorial of 1 and zero is 1 only
# if we initialize factorial by 0 the 0x0 --> 0, then we will get output factorial as zero only.
for i in range(1,num+1):
    factorial *=i

print(f"The Factorial of {num} is {factorial}.")
```
