77. Write a program to Multiply matrices.

```python
rows1 = int(input("Enter number of rows for Matrix 1: "))
cols1 = int(input("Enter number of columns for Matrix 1: "))

rows2 = int(input("\nEnter number of rows for Matrix 2: "))
cols2 = int(input("Enter number of columns for Matrix 2: "))

# Checking if multiplication is possible
if cols1 != rows2:
    print(f"\nMatrix multiplication is not possible because {cols1} != {rows2}.")
else:
    # Input for Matrix 1
    matrix1 = []
    print("\nEnter elements of Matrix 1:")
    for i in range(rows1):
        row = []
        for j in range(cols1):
            row.append(int(input(f"Enter element [{i}][{j}]: ")))
        matrix1.append(row)

    # Input for Matrix 2
    matrix2 = []
    print("\nEnter elements of Matrix 2:")
    for i in range(rows2):
        row = []
        for j in range(cols2):
            row.append(int(input(f"Enter element [{i}][{j}]: ")))
        matrix2.append(row)

    # Printing Matrix 1
    print("\nMatrix 1:")
    for row in matrix1:
        print(" ".join(f"{num}" for num in row))

    # Printing Matrix 2
    print("\nMatrix 2:")
    for row in matrix2:
        print(" ".join(f"{num}" for num in row))

    # Initializing Result Matrix
    result = []
    for i in range(rows1):
        row = []
        for j in range(cols2):
            row.append(0)
        result.append(row)

    # Matrix Multiplication
    for i in range(rows1):
        for j in range(cols2):
            for k in range(cols1):
                result[i][j] += matrix1[i][k] * matrix2[k][j]

    # Printing the Result Matrix
    print("\nResultant Matrix:")
    for row in result:
        print(" ".join(f"{num}" for num in row))
```

78. Write a program to Check symmetric matrix.

```python
n = int(input("Enter the order of the square matrix: "))
matrix = []

print("\nEnter the elements of the matrix:")
for i in range(n):
    row = []
    for j in range(n):
        row.append(int(input(f"Enter element [{i}][{j}]: ")))
    matrix.append(row)

# Printing the matrix
print("\nMatrix:")
for row in matrix:
    print(" ".join(f"{num}" for num in row))

# Checking for the symmetry
is_symmetric = True

for i in range(n):
    for j in range(n):
        if matrix[i][j] != matrix[j][i]:
            is_symmetric = False
            break
    if not is_symmetric:
        break

# printing the result
if is_symmetric:
    print("\nThe matrix is a Symmetric Matrix.")
else:
    print("\nThe matrix is NOT a Symmetric Matrix.")
```

79. Write a program to Find row-wise sum.

```python
rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of columns: "))

matrix = []

print("\nEnter the elements of the matrix:")
for i in range(rows):
    row = []
    for j in range(cols):
        row.append(int(input(f"Enter element [{i}][{j}]: ")))
    matrix.append(row)

# Printing  the matrix
print("\nMatrix:")
for row in matrix:
    print(" ".join(f"{num}" for num in row))

# Finding and printing row-wise sum
print("\nRow-wise Sum:")
for i in range(rows):
    row_sum = 0
    for j in range(cols):
        row_sum += matrix[i][j]
    print(f"Sum of Row {i + 1} = {row_sum}")
```

80. Write a program to Find column-wise sum.

```python
rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of columns: "))

matrix = []

print("\nEnter the elements of the matrix:")
for i in range(rows):
    row = []
    for j in range(cols):
        row.append(int(input(f"Enter element [{i}][{j}]: ")))
    matrix.append(row)

# Printing the matrix
print("\nMatrix:")
for row in matrix:
    print(" ".join(f"{num}" for num in row))

# Finding and printing the  column-wise sum
print("\nColumn-wise Sum:")
for j in range(cols):
    col_sum = 0
    for i in range(rows):
        col_sum += matrix[i][j]
    print(f"Sum of Column {j + 1} = {col_sum}")
```
