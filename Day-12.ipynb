45. Write a program to Write function for palindrome.

```python
def is_palindrome(text):
    return text == text[::-1]

word = input(f"Enter a word: ")

if is_palindrome(word):
    print(f'"{word}" is a palindrome.')
else:
    print(f'"{word}" is not a palindrome.')
```

46. Write a program to Write function for Armstrong.

```python
def is_armstrong(num):
    digits = len(str(num))
    total = sum(int(digit) ** digits for digit in str(num))
    return total == num

number = int(input(f"Enter a number: "))

if is_armstrong(number):
    print(f"{number} is an Armstrong number.")
else:
    print(f"{number} is not an Armstrong number.")
```

47. Write a program to Write function for Fibonacci.

```python
def fibonacci(n):
    a, b = 0, 1

    for _ in range(n):
        print(a, end=" ")
        a, b = b, a + b

n = int(input(f"Enter the number of terms: "))

print(f"Fibonacci series:")
fibonacci(n)
```

48. Write a program to Write function for perfect number.

```python
def is_perfect(num):
    total = 0

    for i in range(1, num):
        if num % i == 0:
            total += i

    return total == num

number = int(input(f"Enter a number: "))

if is_perfect(number):
    print(f"{number} is a Perfect Number.")
else:
    print(f"{number} is not a Perfect Number.")
```
