5. Write a program to Find sum of digits of a number.

```python
num = int(input("Enter the number: ")) # taking input from user for number
temp:int = num # storing original value in temp
sum:int = 0 # initializing sum variable by 0
while(num!=0): # loop will executes unitil num will become equals to zero or less then 1
    d = num%10 # extracting last digit of a number
    sum += d # adding last digit to sum variable
    num = num //10 # excluding the last digit
print(f"The sum of digits of a {temp} is {sum}") 

```

6. Write a program to Reverse a number. 

```python
num = int(input("Enter the number: ")) # taking input from user for number
temp:int = num # storing original value in temp
rev:int = 0 # initializing rev variable by 0
while(num!=0): # loop will executes unitil num will become equals to zero or less then 1
    d = num%10 # extracting last digit of a number
    rev = (rev*10)+d # reversing a number
    num = num //10 # excluding the last digit
print(f"The reverse of of a number {temp} is {rev}") 

```
