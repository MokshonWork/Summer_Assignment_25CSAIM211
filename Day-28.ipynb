109. Write a program to Create library management system.

```python
books = []
count = 0

while True:
    print("\n--- LIBRARY MANAGEMENT SYSTEM ---")
    print("1. Add Book")
    print("2. View Books")
    print("3. Search Book")
    print("4. Issue Book")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        title = input("Enter book title: ")
        author = input("Enter author name: ")

        books.append([title, author, "Available"])
        count = count + 1

        print(f"Book '{title}' added successfully.")

    elif choice == "2":
        if count == 0:
            print("No books in library.")
        else:
            i = 0
            while i < count:
                print(f"{i+1}. {books[i][0]} | {books[i][1]} | {books[i][2]}")
                i = i + 1

    elif choice == "3":
        title = input("Enter book title: ")
        found = False

        i = 0
        while i < count:
            if books[i][0] == title:
                print(f"Title : {books[i][0]}")
                print(f"Author: {books[i][1]}")
                print(f"Status: {books[i][2]}")
                found = True
                break
            i = i + 1

        if not found:
            print("Book not found.")

    elif choice == "4":
        title = input("Enter book title: ")
        found = False

        i = 0
        while i < count:
            if books[i][0] == title:
                found = True
                if books[i][2] == "Available":
                    books[i][2] = "Issued"
                    print("Book issued successfully.")
                else:
                    print("Book already issued.")
                break
            i = i + 1

        if not found:
            print("Book not found.")

    elif choice == "5":
        print("Thank you!")
        break

    else:
        print("Invalid choice.")
```

110. Write a program to Create bank account system.

```python
# Bank Account System

name = input("Enter Account Holder Name: ")
acc_no = input("Enter Account Number: ")
balance = 0.0

while True:
    # Using a multiline f-string to cleanly format the menu without multiple print statements
    menu = f"""
    BANK ACCOUNT SYSTEM 
    1. Deposit
    2. Withdraw
    3. Check Balance
    4. Account Details
    5. Exit
    """
    print(menu)

    choice = input("Enter your choice: ")

    if choice == "1":
        amount = float(input("Enter deposit amount: "))
        balance += amount
        print(f"₹{amount} deposited successfully.")
        print(f"Current Balance: ₹{balance}")

    elif choice == "2":
        amount = float(input("Enter withdrawal amount: "))
        if amount <= balance:
            balance -= amount
            print(f"₹{amount} withdrawn successfully.")
            print(f"Current Balance: ₹{balance}")
        else:
            print("Insufficient balance.")

    elif choice == "3":
        print(f"Current Balance: ₹{balance}")

    elif choice == "4":
        # Using a multiline f-string for beautifully aligned account details
        details = f"""
    ----- Account Details -----
    Account Holder : {name}
    Account Number : {acc_no}
    Balance        : ₹{balance}
    ---------------------------
        """
        print(details)

    elif choice == "5":
        print("Thank you for using the Bank Account System.")
        break

    else:
        print("Invalid choice. Please try again.")
```

111. Write a program to Create ticket booking system.

```python
ticket_price = 200
available_tickets = 50

print("===== Ticket Booking System =====")

name = input("Enter your name: ")
tickets = int(input("Enter number of tickets: "))

if tickets <= 0:
    print("Invalid number of tickets.")
elif tickets > available_tickets:
    print(f"Sorry {name}, only {available_tickets} tickets are available.")
else:
    total_amount = tickets * ticket_price
    available_tickets = available_tickets - tickets

    print("\n----- Booking Details -----")
    print(f"Customer Name : {name}")
    print(f"Tickets Booked: {tickets}")
    print(f"Price per Ticket: ₹{ticket_price}")
    print(f"Total Amount: ₹{total_amount}")
    print(f"Tickets Left: {available_tickets}")
    print("Booking Successful!")
```

112. Write a program to Create contact management system.

```python
name = ""
phone = ""
email = ""

print("===== Contact Management System =====")
print("1. Add Contact")
print("2. View Contact")

choice = int(input("Enter your choice: "))

if choice == 1:
    name = input("Enter Name: ")
    phone = input("Enter Phone Number: ")
    email = input("Enter Email: ")

    print("\nContact Added Successfully!")

elif choice == 2:
    if name == "":
        print("No contact available.")
    else:
        print(f"Name  : {name}")
        print(f"Phone : {phone}")
        print(f"Email : {email}")

else:
    print("Invalid Choice.")
```
