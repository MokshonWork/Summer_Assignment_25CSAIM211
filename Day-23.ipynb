89. Write a program to Find first non-repeating character.

```python
text = input("Enter a string: ")

found = False
print(f"You entered: {text}")

for i in text:
    count = 0

    for j in text:
        if i == j:
            count += 1

    if count == 1:
        print(f"The first non-repeating character is: {i}")
        found = True
        break

if not found:
    print("No non-repeating character found.")
```

90. Write a program to Find first repeating character.

```python
text = input("Enter a string: ")

found = False
print(f"You entered: {text}")
for i in range(len(text)):
    for j in range(i):
        if text[i] == text[j]:
            print(f"The first repeating character is: {text[i]}")
            found = True
            break
    if found:
        break

if not found:
    print("No repeating character found.")
```

91. Write a program to Check anagram strings.

```python
str1 = input("Enter the first string: ")
str2 = input("Enter the second string: ")

if len(str1) != len(str2):
    print(f'"{str1}" and "{str2}" are not anagrams.')
else:
    is_anagram = True

    for ch in str1:
        count1 = 0
        count2 = 0

        for i in str1:
            if i == ch:
                count1 += 1

        for j in str2:
            if j == ch:
                count2 += 1

        if count1 != count2:
            is_anagram = False
            break

    if is_anagram:
        print(f'"{str1}" and "{str2}" are anagrams.')
    else:
        print(f'"{str1}" and "{str2}" are not anagrams.')
```

92. Write a program to Find maximum occurring character.

```python
text = input("Enter a string: ")

max_count = 0
max_char = ""

for i in text:
    count = 0

    for j in text:
        if i == j:
            count += 1

    if count > max_count:
        max_count = count
        max_char = i

print(f"The maximum occurring character is: {max_char}")
print(f"It occurs {max_count} times.")
```
