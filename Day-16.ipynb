61. Write a program to Find missing number in array.


```python
def find_missing_number(arr, n):
    expected_sum = n * (n + 1) // 2

    actual_sum = 0
    for num in arr:
        actual_sum += num

    return expected_sum - actual_sum

arr = [1, 2, 4, 5, 6]
n = 6
missing = find_missing_number(arr, n)
print(f"Missing Number: {missing}")
```

62. Write a program to Find maximum frequency element.

```python
def max_frequency_element(arr):
    max_count = 0
    max_element = arr[0]

    for i in range(len(arr)):
        count = 0

        for j in range(len(arr)):
            if arr[i] == arr[j]:
                count += 1

        if count > max_count:
            max_count = count
            max_element = arr[i]

    return max_element, max_count

arr = [1, 2, 2, 3, 4, 2, 5, 3, 3, 3]

element, frequency = max_frequency_element(arr)
print(f"Maximum Frequency Element: {element}")
print(f"Frequency: {frequency}")
```

63. Write a program to Find pair with given sum.

```python
def find_pair_with_sum(arr, target):
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            if arr[i] + arr[j] == target:
                return arr[i], arr[j]

    return None

arr = [1, 4, 5, 7, 3, 9]
target = 10

pair = find_pair_with_sum(arr, target)

if pair:
    print(f"Pair found: {pair[0]} and {pair[1]}")
else:
    print(f"No pair found with sum {target}")
```

64. Write a program to Remove duplicates from array.

```python
def remove_duplicates(arr):
    unique = []

    for num in arr:
        if num not in unique:
            unique.append(num)

    return unique

arr = [1, 2, 2, 3, 4, 4, 5, 1]
print(f"Array after removing duplicates: {remove_duplicates(arr)}")
```
