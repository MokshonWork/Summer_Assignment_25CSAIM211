105. Write a program to Create student record management system.

```python
students = []

while True:
    print("\n-- Student Record Management System ---")
    print("1. Add Student")
    print("2. View Students")
    print("3. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        name = input("Enter student name: ")
        roll = input("Enter roll number: ")
        marks = float(input("Enter marks: "))

        students.append([name, roll, marks])

        print(f"Student '{name}' added successfully!")

    elif choice == 2:
        if len(students) == 0:
            print("No student records found.")
        else:
            print("\n----- Student Records -----")
            for student in students:
                print(f"Name: {student[0]}")
                print(f"Roll No: {student[1]}")
                print(f"Marks: {student[2]}")
                print("---------------------------")

    elif choice == 3:
        print("Exiting Student Record Management System.")
        break

    else:
        print("Invalid choice! Please try again.")
```

106. Write a program to Create employee management system.


```python
employees = []

while True:
    print("\n-- Employee Management System ---")
    print("1. Add Employee")
    print("2. View Employees")
    print("3. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        emp_id = input("Enter Employee ID: ")
        name = input("Enter Employee Name: ")
        department = input("Enter Department: ")
        salary = float(input("Enter Salary: "))

        employees.append([emp_id, name, department, salary])

        print(f"Employee '{name}' added successfully!")

    elif choice == 2:
        if len(employees) == 0:
            print("No employee records found.")
        else:
            print("\n----- Employee Records -----")
            for employee in employees:
                print(f"Employee ID: {employee[0]}")
                print(f"Name: {employee[1]}")
                print(f"Department: {employee[2]}")
                print(f"Salary: {employee[3]}")
                print("----------------------------")

    elif choice == 3:
        print("Exiting Employee Management System.")
        break

    else:
        print("Invalid choice! Please try again.")
```

107. Write a program to Create salary management system.

```python
employees = []

while True:
    print("\n--- Salary Management System ---")
    print("1. Add Employee Salary")
    print("2. View Employee Salaries")
    print("3. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        emp_id = input("Enter Employee ID: ")
        name = input("Enter Employee Name: ")
        salary = float(input("Enter Salary: "))

        employees.append([emp_id, name, salary])

        print(f"Salary record for {name} added successfully!")

    elif choice == 2:
        if len(employees) == 0:
            print("No salary records found.")
        else:
            print("\n----- Salary Records -----")
            for employee in employees:
                print(f"Employee ID : {employee[0]}")
                print(f"Name        : {employee[1]}")
                print(f"Salary      : {employee[2]}")
                print("--------------------------")

    elif choice == 3:
        print("Exiting Salary Management System.")
        break

    else:
        print("Invalid choice! Please try again.")
```

108. Write a program to Create marksheet generation system.


```python
name = input("Enter student name: ")
roll = input("Enter roll number: ")

math = float(input("Enter Math marks: "))
science = float(input("Enter Science marks: "))
english = float(input("Enter English marks: "))

total = math + science + english
percentage = total / 3

if percentage >= 90:
    grade = "A+"
elif percentage >= 80:
    grade = "A"
elif percentage >= 70:
    grade = "B"
elif percentage >= 60:
    grade = "C"
elif percentage >= 50:
    grade = "D"
else:
    grade = "F"

print("\n===== Student Marksheet =====")
print(f"Name       : {name}")
print(f"Roll No    : {roll}")
print(f"Math       : {math}")
print(f"Science    : {science}")
print(f"English    : {english}")
print(f"Total      : {total}")
print(f"Percentage : {percentage:.2f}%")
print(f"Grade      : {grade}")

if grade == "F":
    print("Result     : Fail")
else:
    print("Result     : Pass")
```
