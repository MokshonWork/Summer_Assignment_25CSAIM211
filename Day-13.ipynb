49. Write a program to Input and display array.

```python
n = int(input("Enter number of elements: "))#taking input size of array from the user
# Inputd elements from user
arr = []
for i in range(n):
    element = int(input(f"Enter element {i + 1}: "))
    arr.append(element)

# printing array elements from user
print("Array elements are:")
for item in arr:
    print(item, end=" ")
```

50. Write a program to Find sum and average of array.

```python
n = int(input("Enter number of elements: "))#taking input size of array from the user
# Inputed array elements from user
arr = []
for i in range(n):
    arr.append(int(input(f"Enter element {i+1}: ")))

# Calculating sum and average
total = sum(arr)
average = total / n

#results
print("Array:", arr)
print("Sum =", total)
print("Average =", average)
```

51. Write a program to Find largest and smallest element.

```python
n = int(input("Enter number of elements: "))#taking input size of array from the user
# Inputed array elements from user
arr = []
for i in range(n):
    arr.append(int(input(f"Enter element {i+1}: ")))

# printing array elements from user
print("Array elements are:")
for item in arr:
    print(item, end=" ")

largest = arr[0]
smallest = arr[0]

for num in arr:
    if num > largest:
        largest = num
    if num < smallest:
        smallest = num

print(" ")
print("Largest element =", largest)
print("Smallest element =", smallest)
```

52. Write a program to Count even and odd elements.

```python
n = int(input("Enter number of elements: "))#taking input size of array from the user
# Inputed array elements from user
arr = []
for i in range(n):
    arr.append(int(input(f"Enter element {i+1}: ")))

# printing array elements from user
print("Array elements are:")
for item in arr:
    print(item, end=" ")

even_count = 0
odd_count = 0

for num in arr:
    if num % 2 == 0:
        even_count += 1
    else:
        odd_count += 1

print(" ")
print("Number of even elements =", even_count)
print("Number of odd elements =", odd_count)
```
