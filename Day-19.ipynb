73. Write a program to Add matrices.

```python
A = [[1, 2],
     [3, 4]]

B = [[5, 6],
     [7, 8]]

C = []

for i in range(2):
    row = []
    for j in range(2):
        row.append(A[i][j] + B[i][j])
    C.append(row)

print("Sum of matrices:")
for i in range(2):
    for j in range(2):
        print(f"{C[i][j]}", end=" ")
    print()
```

74. Write a program to Subtract matrices.

```python
# Matrix Subtraction

A = [[5, 6],
     [7, 8]]

B = [[1, 2],
     [3, 4]]

C = []

for i in range(2):
    row = []
    for j in range(2):
        row.append(A[i][j] - B[i][j])
    C.append(row)

print("Subtraction of matrices:")
for i in range(2):
    for j in range(2):
        print(f"{C[i][j]}", end=" ")
    print()
```

75. Write a program to Transpose matrix.

```python
r = int(input("Enter number of rows: "))
c = int(input("Enter number of columns: "))

A = []

print("Enter matrix elements:")
for i in range(r):
    row = []
    for j in range(c):
        num = int(input(f"Enter element A[{i}][{j}]: "))
        row.append(num)
    A.append(row)

print("\nInput Matrix:")
for i in range(r):
    for j in range(c):
        print(f"{A[i][j]}", end=" ")
    print()

print("\nTranspose of the Matrix:")
for i in range(c):
    for j in range(r):
        print(f"{A[j][i]}", end=" ")
    print()
```

76. Write a program to Find diagonal sum.

```python
n = int(input("Enter the size of the square matrix: "))

A = []

print("Enter matrix elements:")
for i in range(n):
    row = []
    for j in range(n):
        num = int(input(f"Enter element A[{i}][{j}]: "))
        row.append(num)
    A.append(row)

print("\nInput Matrix:")
for i in range(n):
    for j in range(n):
        print(f"{A[i][j]}", end=" ")
    print()

sum = 0

for i in range(n):
    sum = sum + A[i][i]

print(f"\nDiagonal Sum = {sum}")
```
