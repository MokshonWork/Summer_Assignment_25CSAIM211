17. Write a program to Check perfect number.

```python
num = int(input("Enter a number: "))#--> input from user

sum_divisors = 0 #initializing sum_divisors as 0

for i in range(1, num):#iterating loop from 1-->num-1
    if num % i == 0:#checking condition
        sum_divisors += i#addimg sum_divisor with i,
        #which is divisible by n

#printing perfect number or not
if sum_divisors == num:
    print(num, "is a Perfect Number")
else:
    print(num, "is not a Perfect Number")
```

18. Write a program to Check strong number.

```python
num = int(input("Enter a number: "))#--> input from user
temp = num #storing original value in temp 
sum_fact = 0 #intializing as 0

while num > 0:#loop will iterate unil condition becomes false
    digit = num % 10

    fact = 1
    for i in range(1, digit + 1):#fatorial
        fact *= i

    sum_fact += fact #logic
    num //= 10
#printing strong no. or not
if sum_fact == temp:
    print(temp, "is a Strong Number")
else:
    print(145, "is not a Strong Number")
```

19. Write a program to Print factors of a number.

```python
num = int(input("Enter a number: "))#--> input from user
print(f"Factors of{num} are:")

for i in range(1, num + 1):#loop iterating from 1  --> n
    if num % i == 0:#checking condition
        print(i, end=" ")
```
