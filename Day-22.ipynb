85. Write a program to Check palindrome string.

```python
text = input("Enter a string: ")

reverse = ""

for ch in text:
    reverse = ch + reverse

print(f"You entered: {text}")

if text == reverse:
    print("The string is a palindrome.")
else:
    print("The string is not a palindrome.")
```

86. Write a program to Count words in a sentence.

```python
text = input("Enter a sentence: ")

count = 1

for ch in text:
    if ch == ' ':
        count += 1

print(f"You entered: {text}")
print(f"Number of words: {count}")
```

87. Write a program to Character frequency.


```python
text = input("Enter a string: ")
ch = input("Enter the character to find: ")

count = 0

for i in text:
    if i == ch:
        count += 1

print(f"You entered: {text}")
print(f"Frequency of '{ch}' is: {count}")
```

88. Write a program to Remove spaces from a string.


```python
text = input("Enter a string: ")

result = ""

for ch in text:
    if ch != ' ':
        result = result + ch

print(f"You entered: {text}")
print(f"String without spaces: {result}")
```
