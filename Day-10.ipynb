37. Write a program to Print star pyramid.
```text
    *
   ***
  *****
 *******
*********
```

````python
rows = 5

for i in range(1, rows + 1):
    spaces = rows - i
    stars = 2 * i - 1
    print(" " * spaces + "*" * stars)
````

38. Write a program to Print reverse pyramid.
```text
*********
 *******
  *****
   ***
    *
```

````python
rows = 5

for i in range(rows, 0, -1):
    spaces = rows - i
    stars = 2 * i - 1
    print(" " * spaces + "*" * stars)
````

39. Write a program to Print number pyramid.
```text
    1
   121
  12321
 1234321
123454321
```

````python
rows = 5

for i in range(1, rows + 1):
    print(" " * (rows - i), end="")

    for j in range(1, i + 1):
        print(j, end="")

    for j in range(i - 1, 0, -1):
        print(j, end="")

    print()
````

40. Write a program to Print character pyramid.
```text
    A
   ABA
  ABCBA
 ABCDCBA
ABCDEDCBA
```


````python
rows = 5

for i in range(1, rows + 1):
    print(" " * (rows - i), end="")

    for j in range(i):
        print(chr(65 + j), end="")

    for j in range(i - 2, -1, -1):
        print(chr(65 + j), end="")

    print()
````
