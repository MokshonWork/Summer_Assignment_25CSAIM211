9. Write a program to Check whether a number is prime.

```python
num = int(input("Enter the number: ")) # taking input from the user

is_prime = True # assuming give no. is prime

for i in range(2, num): # loop running from 2 to num -1
    # If num is divisible by i, it's not prime
    if num % i == 0:
        is_prime = False
        break  

if is_prime and num > 1: # A prime number must be greater than 1
    print(f"{num} is a prime number")
else:
    print(f"{num} is not a prime number")
```

10. Write a program to Print prime numbers in a range.10 

```python
start = 1#starting point
end = 10#ending point

for i in range(start, end + 1):
    if i > 1:
        is_prime = True
        
        for n in range(2, int(i**0.5) + 1):
            if i % n == 0:
                is_prime = False
                break

        # printing the numbers
        if is_prime:
            print(i)
```

11. Write a program to Find GCD of two numbers.

```python
# taking two numbers as the input from the user
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

# Storing the greatest common divisor in the variable name gcd
gcd = 1

# Checking all possible common factors
for i in range(1, min(num1, num2) + 1):
    # If i is dividing both numbers, then updating the gcd
    if num1 % i == 0 and num2 % i == 0:
        gcd = i

print(f"The GCD of {num1} and {num2} is {gcd}")# printing the gcd
```

12. Write a program to Find LCM of two numbers.

```python
# taking two numbers as input from the user
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

largest = max(num1, num2)#checking from the larger number

# searching until a common multiple is found
while True:
    # until largest is divisible is by both numbers, it is the going to be LCM
    if largest % num1 == 0 and largest % num2 == 0:
        lcm = largest
        break
    largest += 1 #incrementing by 1

print(f"The  of LCM Of{num1} and {num2} is {lcm}")# printing the lcm
```
