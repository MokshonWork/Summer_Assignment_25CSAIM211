69. Write a program to Bubble sort.


```python
nums = [5,7,8,4,1,6,9,2 ]

def bubble_sort(data: list[int]) -> None:
    n = len(data)
    for i in range(n-2,-1,-1):
        for j in range(0,i+1):
            if data[j] > data[j+1]:
                data[j],data[j+1] = data[j+1],data[j]

bubble_sort(nums)
print(nums)
```

70. Write a program to Selection sort.

```python
nums = [5,7,8,4,1,6,9,2 ]

def selection_sort(data: list[int]) -> None:
    n = len(data)
    for i in range(0,n):
        min_index =i
        for j in range(i+1,n):
            if data[j] < data[min_index]:
                min_index = j
        
        data[i],data[min_index] = data[min_index],data[i]

selection_sort(nums)
print(nums)
```

71. Write a program to Binary search.


```python
def binary_search(arr, key):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == key:

            return mid
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1

    return -1

arr = list(map(int, input("Enter sorted numbers separated by space: ").split()))
key = int(input("Enter the element to search: "))
print(arr)
result = binary_search(arr, key)
print(" ")
if result != -1:
    print(f"Element {key} found at index {result}.")
else:
    print(f"Element {key} not found in the list.")
```

72. Write a program to Sort array in descending order.

```python
def descending_sort(arr):
    n = len(arr)

    for i in range(n):
        for j in range(n - i - 1):
            if arr[j] < arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

    return arr

arr = list(map(int, input("Enter array elements: ").split()))
print(arr)
result = descending_sort(arr)
print(" ")
print(f"Array in descending order: {result}")
```
