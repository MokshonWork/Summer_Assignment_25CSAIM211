29. Write a program to Print half pyramid pattern.

````python
rows = int(input("Enter number of rows: "))# Take number of rows
# Print half pyramid
for i in range(1, rows + 1):
    print("*" * i)
````

30. Write a program to Print number triangle.

```text
1
12
123
1234
12345
```



````python
# Take number of rows
rows = int(input("Enter number of rows: "))
# Print number triangle
for i in range(1, rows + 1):
    for j in range(1, i + 1):
        print(j, end="")
    print()
````

31. Write a program to Print character triangle.
```text
A
AB
ABC
ABCD
ABCDE
```


````python
# Take input from user
rows = int(input("Enter number of rows: "))

# Print the pattern
for i in range(1, rows + 1):
    for j in range(i):
        print(chr(65 + j), end="")
    print()
````

32. Write a program to Print repeated-number pattern.
```text
1
22
333
4444
55555
```



````python
# Take number of rows
rows = int(input("Enter number of rows: "))

for i in range(1, rows + 1):
    print(str(i) * i)
````
