57. Write a program to Reverse array.

```python
arr = [1, 2, 3, 4, 5]

left = 0
right = len(arr) - 1

while left < right:
    arr[left], arr[right] = arr[right], arr[left]
    left += 1
    right -= 1

print(f"Reversed array:{arr}")
```

```python
arr = [1, 2, 3, 4, 5]

reversed_arr = arr[::-1]

print(f"Reversed array:{reversed_arr}")
```

58. Write a program to Rotate array left.

```python
arr = [1, 2, 3, 4, 5]
k = 2

n = len(arr)

for _ in range(k):
    first = arr[0]

    for i in range(n - 1):
        arr[i] = arr[i + 1]

    arr[n - 1] = first

print(f"Left rotated array:{arr}")
```

```python
arr = [1, 2, 3, 4, 5]
k = 2

rotated = arr[k:] + arr[:k]

print(f"Left rotated array:{rotated}")
```

59. Write a program to Rotate array right.

```python
arr = [1, 2, 3, 4, 5]
k = 2

n = len(arr)

for _ in range(k):
    last = arr[n - 1]

    for i in range(n - 1, 0, -1):
        arr[i] = arr[i - 1]

    arr[0] = last

print(f"Right rotated array:{arr}")
```

60. Write a program to Move zeroes to end.

```python
arr = [0, 1, 0, 3, 12]

result = []

# Add non-zero elements
for num in arr:
    if num != 0:
        result.append(num)

# Count and append zeroes
zero_count = len(arr) - len(result)

for _ in range(zero_count):
    result.append(0)

print(f"Array after moving zeroes to end:{result}")
```

```python
arr = [0, 1, 0, 3, 12]

pos = 0

for i in range(len(arr)):
    if arr[i] != 0:
        arr[pos], arr[i] = arr[i], arr[pos]
        pos += 1

print(f"Array after moving zeroes to end:{arr}")
```
