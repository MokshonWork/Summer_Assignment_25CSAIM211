53. Write a program to Linear search.

```python
def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i  # Return index if found
    return -1  # Return -1 if not found

# Input list
arr = [10, 20, 30, 40, 50]

# Element to search
key = int(input("Enter the element to search: "))

# Call function
result = linear_search(arr, key)

if result != -1:
    print(f"Element found at index {result}")
else:
    print("Element not found")
```

54. Write a program to Frequency of an element.

```python
n = int(input("Enter number of elements: "))
arr = []

for i in range(n):
    arr.append(int(input("Enter element: ")))

# printing array elements from user
print("Array elements are:")
for item in arr:
    print(item, end=" ")

element = int(input("Enter the element to find frequency: "))

count = arr.count(element)
print(" ")
print(f"Frequency of {element} is {count}")
```

55. Write a program to Second largest element.

```python
def se_l(data):
    largest = float("-inf")
    sec_largest = float("-inf")
    n = len(data)

    for i in range(0,n):
        if data[i] > largest:
            sec_largest = largest
            largest = data[i]
        elif data[i] > sec_largest and data[i] != largest:
            sec_largest = data[i]

    return sec_largest
        
nums = [55, 32, 97, -55, 45, 32, 88, 21]
print(se_l(nums))
```

56. Write a program to Find duplicates in array.

```python
n = int(input("Enter number of elements: "))
arr = []

for i in range(n):
    arr.append(int(input("Enter element: ")))
print("Array elements are:")
for item in arr:
    print(item, end=" ")

duplicates = []

for i in range(len(arr)):
    for j in range(i + 1, len(arr)):
        if arr[i] == arr[j] and arr[i] not in duplicates:
            duplicates.append(arr[i])

print(" ")
print(f"Duplicate elements are: {duplicates}")
```
