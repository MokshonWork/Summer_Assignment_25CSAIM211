117. Write a program to Create student record system using arrays and strings.

```python
names = []
marks = []

n = int(input("Enter the number of students: "))

for i in range(n):
    name = input(f"Enter name of student {i + 1}: ")
    mark = int(input(f"Enter marks of {name}: "))

    names.append(name)
    marks.append(mark)

print("\n----- Student Records -----")
for i in range(len(names)):
    print(f"Student {i + 1}: Name = {names[i]}, Marks = {marks[i]}")

average = sum(marks) / len(marks)

print("\n----- Statistics -----")
print(f"Total Students : {len(names)}")
print(f"Highest Marks  : {max(marks)}")
print(f"Lowest Marks   : {min(marks)}")
print(f"Average Marks  : {average:.2f}")

search = input("\nEnter student name to search: ")

if search in names:
    index = names.index(search)
    print(f"{search} scored {marks[index]} marks.")
else:
    print(f"Student '{search}' not found.")
```

118. Write a program to Create mini library system.

```python
books = []
n = int(input("Enter the number of books: "))

for i in range(n):
    book = input(f"Enter book {i + 1} name: ")
    books.append(book)

print("\n----- Library Books -----")
for i in range(len(books)):
    print(f"{i + 1}. {books[i]}")

search = input("\nEnter book name to search: ")

if search in books:
    print(f'"{search}" is available in the library.')
else:
    print(f'"{search}" is not available.')

remove_book = input("\nEnter book name to remove: ")

if remove_book in books:
    books.remove(remove_book)
    print(f'"{remove_book}" has been removed.')
else:
    print(f'"{remove_book}" not found.')

print("\n----- Updated Library -----")
for i in range(len(books)):
    print(f"{i + 1}. {books[i]}")

print(f"\nTotal books available: {len(books)}")
```

119. Write a program to Create mini employee management system.

```python
names = []
salaries = []

n = int(input("Enter the number of employees: "))

for i in range(n):
    name = input(f"Enter employee {i + 1} name: ")
    salary = float(input(f"Enter salary of {name}: "))

    names.append(name)
    salaries.append(salary)

print("\n----- Employee Records -----")
for i in range(len(names)):
    print(f"Employee {i + 1}: Name = {names[i]}, Salary = ${salaries[i]:.2f}")

search = input("\nEnter employee name to search: ")

if search in names:
    index = names.index(search)
    print(f"{search}'s Salary = ${salaries[index]:.2f}")
else:
    print("Employee not found.")

print("\n----- Salary Statistics -----")
print(f"Total Employees : {len(names)}")
print(f"Highest Salary  : ${max(salaries):.2f}")
print(f"Lowest Salary   : ${min(salaries):.2f}")
print(f"Average Salary  : ${sum(salaries) / len(salaries):.2f}")
```

120. Write a program to Develop complete mini project using arrays, strings and functions.

```python
names = []
marks = []

def add_student():
    name = input("Enter student name: ")
    mark = int(input("Enter marks: "))
    names.append(name)
    marks.append(mark)
    print(f"{name} added successfully.\n")

def display_students():
    if len(names) == 0:
        print("No student records found.\n")
    else:
        print("\n----- Student Records -----")
        for i in range(len(names)):
            print(f"{i + 1}. Name: {names[i]}, Marks: {marks[i]}")
        print()

def search_student():
    name = input("Enter student name to search: ")
    if name in names:
        index = names.index(name)
        print(f"{name} scored {marks[index]} marks.\n")
    else:
        print("Student not found.\n")

def statistics():
    if len(marks) == 0:
        print("No records available.\n")
    else:
        print("\n----- Statistics -----")
        print(f"Total Students : {len(names)}")
        print(f"Highest Marks  : {max(marks)}")
        print(f"Lowest Marks   : {min(marks)}")
        print(f"Average Marks  : {sum(marks)/len(marks):.2f}\n")

while True:
    print("===== Student Management System =====")
    print("1. Add Student")
    print("2. Display Students")
    print("3. Search Student")
    print("4. Statistics")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_student()
    elif choice == "2":
        display_students()
    elif choice == "3":
        search_student()
    elif choice == "4":
        statistics()
    elif choice == "5":
        print("Thank you! Exiting program.")
        break
    else:
        print("Invalid choice. Try again.\n")
```
